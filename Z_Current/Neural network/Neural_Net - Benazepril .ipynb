### Read datasets

```python
PCA_df = pd.read_csv(r"Sets/pca_3.csv",index_col=[0])
PCA_df
```

```python
df = pd.read_csv(r"Sets/drug_onehot_latest.csv")
df
```

### Run the Network

```python
q = 'Benazepril hydrochloride tablet'
```

```python
print("Running for :",q)
trainer = pd.concat([PCA_df, df[q]], axis = 1)
```

```python
# trainer
```

```python
X = trainer.drop(q,axis='columns')
y = trainer[q]
print(y.value_counts())
smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)
print('SMOTE')
print(q)
print(y_sm.value_counts().sum())


# print('Random over-sampling:')
# print(df_test_over[drug].value_counts())
# X = df_test_over.drop(drug,axis='columns')
# y = df_test_over[drug]

X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)
loss = keras.losses.BinaryCrossentropy()
weights = -1


```

```python
# from keras.callbacks import History 
# history = History()
```

```python
# from keras.callbacks import EarlyStopping
# early_stopping = EarlyStopping(monitor='val_loss', patience=4, mode='auto')
```

```python
reg_model = Sequential()
reg_model.add(Dense(32, input_dim=46, activation='relu', kernel_regularizer='l2'))
# reg_model.add(Dense(128, activation='relu'))
reg_model.add(Dropout(0.2))
# reg_model.add(Dropout(0.2))
reg_model.add(Dense(16, activation='relu'))
reg_model.add(Dropout(0.2))
#reg_model.add(Dense(32, activation='relu'))
reg_model.add(Dense(1, activation='sigmoid'))
reg_model.compile(loss='binary_crossentropy', 
                optimizer='adam', 
                metrics=['accuracy'])


history = reg_model.fit(X_train, y_train, 
                            validation_data=(X_test, y_test), 
                            epochs=100, verbose=0)

print(reg_model.evaluate(X_test, y_test))

y_preds = reg_model.predict(X_test)
y_preds = np.round(y_preds)

print("Classification Report: \n", classification_report(y_test, y_preds))
```

### Plot characteristic curves

```python
plt.figure(figsize=(5,4),dpi=150)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend()
plt.title('Loss Curve for Benazepril hydrochloride tablet')
plt.ylim(0.2,1.2)
plt.show()
```

```python
#Predict on test set
predictions_NN_prob = reg_model.predict(X_test)
predictions_NN_prob = predictions_NN_prob[:,0]

predictions_NN_01 = np.where(predictions_NN_prob > 0.5, 1, 0)
```

```python
#Print accuracy
acc_NN = accuracy_score(y_test, predictions_NN_01)
print('Overall accuracy of Neural Network model:', acc_NN)
```

```python
false_positive_rate, recall, thresholds = roc_curve(y_test, predictions_NN_prob)
roc_auc = auc(false_positive_rate, recall)
plt.figure(dpi=150)
plt.title('Receiver Operating Characteristic of Benazepril hydrochloride tablet(ROC)')
plt.plot(false_positive_rate, recall, 'b', label = 'AUC = %0.3f' %roc_auc)
plt.legend(loc='lower right')
plt.plot([0,1], [0,1], 'r--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.ylabel('Recall')
plt.xlabel('Fall-out (1-Specificity)')
plt.show()
```

```python
cm = confusion_matrix(y_test, predictions_NN_01)
labels = ['No Default', 'Default']
plt.figure(figsize=(5,4),dpi=150)
sns.heatmap(cm,xticklabels=labels, yticklabels=labels, annot=True, fmt='d', cmap="Blues", vmin = 0.2);
plt.title('Confusion Matrix of Benazepril hydrochloride tablet')
plt.ylabel('True Class')
plt.xlabel('Predicted Class')
plt.show()
```
