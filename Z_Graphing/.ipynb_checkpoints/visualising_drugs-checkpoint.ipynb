```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"drug_onehot_latest.csv",index_col = [0])
```

```python
x = df.columns
```

```python
count_0 , count_1 = df['Digoxin tablet'].value_counts()[0] ,df['Digoxin tablet'].value_counts()[1]
# print(count_0, count_1)
```

```python
a = 0
b = a +1001.5
```

```python
# plt.figure(figsize=(1000, 1000))
plt.rcParams["figure.figsize"] = (100, 50)
```

```python
#plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
```

```python
c = 18
for q in x:
    count_0 , count_1 = df[q].value_counts()[0] ,df[q].value_counts()[1]
    if c >1:
        plt.bar(q, count_0, color='#79DAE8', align = 'edge')
        plt.bar(q, count_1, bottom=count_0, color='#4B7BE5', align = 'edge')
        c-=1
    else:
        plt.bar(q, count_0, color='#79DAE8', align = 'edge',label ="Zeroes")
        plt.bar(q, count_1, bottom=count_0, color='#4B7BE5', align = 'edge',label ="Ones")
    #plt.bar(fontsize = 30)
    plt.xticks(fontsize = 30,rotation = 45)
    plt.yticks(fontsize = 40)
    plt.axhline(y=1001.5, xmin=0, xmax=1, color='#B20600', linestyle='-.', linewidth=10)
    plt.legend(["Zero","one"])
    # plt.setp(ax.get_xticklabels(), rotation=30, horizontalalignment='right')
plt.figure(facecolor='#FF8C32')
```

```python
plt.show()
```
