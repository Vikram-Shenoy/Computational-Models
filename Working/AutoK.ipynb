## AutoKeras

### Initialising data

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
import autokeras as ak
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv("random_imputed.csv",index_col=[0])
df1.sample(5)
```

```python
df3=pd.read_csv("drug_onehot_latest.csv",index_col=[0])
df3.head()
```

```python
df2=df3['Atorvastatin calcium tablet']
df2 = df2.to_frame()
```

```python
df2.head()
```

```python
cols_to_scale = [df1.columns]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_df= scaler.fit_transform(df1)
```

```python
df = pd.DataFrame(scaled_df,columns = df1.columns)
```

```python
df.head()
```

### First model
* No preprocessing or sampling (Accuracy:60%)

```python
X = df
y = df2
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
model_search = ak.StructuredDataClassifier(max_trials=20)
```

```python
model_search.fit(x=X_train, y = y_train)
```

```python
model_search.fit(x = X_train, y = y_train)
```

```python
loss , acc = model_search.evaluate(x = X_train, y = y_train, verbose= 0)
print(f"Train Accuracy = {acc}")
```

```python
loss , acc = model_search.evaluate(x = X_test, y = y_test, verbose= 0)
print(f"Test Accuracy = {acc}")
```
