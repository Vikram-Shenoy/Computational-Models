```python
df3=df2['Atorvastatin calcium tablet']
df3 = df3.to_frame()
```

```python
df3.head()
```

```python
#Lets try and train the model against just one tablet
#i will use XGBoost that tests the best way of imputation and does it 
```

```python
import xgboost as xgb
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1, df3,test_size=0.3, random_state=1121218)
```

```python
xgb_cl= xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
```

```python
xgb_cl.fit(X_train, y_train)
```

```python
preds = xgb_cl.predict(X_test)
```

```python
accuracy_score(y_test, preds)
```

```python

param_grid = {
    "max_depth": [3, 4, 5, 7],
    "learning_rate": [0.1, 0.01, 0.05],
    "gamma": [0, 0.25, 1],
    "reg_lambda": [0, 1, 10],
    "scale_pos_weight": [1, 3, 5],
    "subsample": [0.8],
    "colsample_bytree": [0.5],
}
```

```python
from sklearn.model_selection import GridSearchCV

# Init classifier
xgb_cl = xgb.XGBClassifier(objective="binary:logistic")

# Init Grid Search
grid_cv = GridSearchCV(xgb_cl,param_grid, n_jobs=-1, cv=3, scoring="roc_auc")

# Fit
_ = grid_cv.fit(X_train, y_train)
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
#Some parameters are at the end of their ranges meaning that we have to keep exploring:
```

```python
# Insert the new fixed values to the grid
param_grid["scale_pos_weight"] = [3]
param_grid["subsample"] = [0.8]
param_grid["colsample_bytree"] = [0.5]
param_grid["gamma"] = [0.25]

# Give new value ranges to other params
param_grid["max_depth"] = [1,2,3]
param_grid["reg_lambda"] = [10, 30, 50]
param_grid["learning_rate"] = [0.2,0.3, 0.5, 0.7, 1]
```

```python
grid_cv_2 = GridSearchCV(xgb_cl, param_grid, 
                         cv=3, scoring="roc_auc", n_jobs=-1)

_ = grid_cv_2.fit(X_train, y_train)
```
